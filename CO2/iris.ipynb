```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import pandas as pd
pf= pd.read_csv("iris.csv")
print("\n________________INFO_______________")
pf.info()
print("\n________________HEAD_______________")
pf.head()

```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import ConfusionMatrixDisplay

x=pf.iloc[:,0:3].values
print(x)
y=pf.iloc[:,4].values
print(y)

le = LabelEncoder()
y_new = le.fit_transform(y)
print(y_new)

x_train,x_test,y_train,y_test = train_test_split(x,y_new,test_size=.25)


print("\n________________X TEST_______________")
print(x_test)
print("\n________________X TRAIN______________")
print(x_train)
print("\n________________Y TEST_______________")
print(y_test)
print("\n________________Y TRAIN_______________")
print(y_train)

pf.info()
```

```python
knn = KNeighborsClassifier(n_neighbors =5)
knn.fit(x_train,y_train)
y_predict = knn.predict(x_test)
print("\n\n-------Actual class------\n",y_test)
print("\n------------Predict Class---------\n",y_predict)
print(accuracy_score(y_test,y_predict))

cm=confusion_matrix(y_test,y_predict)
display=ConfusionMatrixDisplay(confusion_matrix=cm)
display.plot()
plt.show()
```
